```python
# minimal imports for biopython
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.Alphabet import IUPAC
from Bio.SeqRecord import SeqRecord
```

```python
probe_file = os.path.join(library_folder, 'final_probes', 'filtered_probes.fasta')
```

```python
with open(probe_file, "r") as handle:
    probes = []
    for record in SeqIO.parse(handle, "fasta"):
        probes.append(record);
```

```python
region_dict = {}
combo_bcs = []
unique_bcs = []

for _pb in probes:
    _id = _pb.id
    _reg = int(_id.split('reg_')[1].split('_gene')[0])
    if _reg not in region_dict:
        _readouts = _id.split('barcodes_')[1].split('_color')[0]
        _combo = _readouts.split(']')[0].split('[')[-1].split(',')
        _combo = [_bc[1:-1] for _bc in _combo]
        for _b in _combo:
            if _b not in combo_bcs:
                combo_bcs.append(_b)
        _unique = _readouts.split(']')[1].split(',')[1][1:-1]
        if _unique not in unique_bcs:
            unique_bcs.append(_unique)
        _color = int(_id.split('color_')[-1])
        _info_dict = {'reg': _reg,
                      'color': _color,
                      'combo': _combo,
                      'unique': _unique}
        region_dict[_reg] = _info_dict
print(len(region_dict))
```

```python
combo_bcs = sorted(combo_bcs, key=lambda v:int(v.split('_')[1]))
combo_groups = [combo_bcs[5*i:5*i+5] for i in range(int(len(combo_bcs)/5))]
```

```python
# find unique_groups
unique_groups = []
for _group in combo_groups:
    _unique_group = []
    for _r, _info in region_dict.items():
        _cb = _info['combo']
        for _b in _cb:
            if _b in _group and _info['unique'] not in _unique_group:
                _unique_group.append(_info['unique'])
    _unique_group = sorted(_unique_group, key=lambda v:int(v.split('_')[1]))
    unique_groups.append(_unique_group)
```
